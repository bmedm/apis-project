##### Estudiaremos primero cuantas peliculas ha realizado cada director, y la media de puntuación de las peliculas de cada uno.

```python
plt.subplots(figsize=(20,5))
sns.countplot('director',data=ghibli,palette="BuGn_r",edgecolor=sns.color_palette('cubehelix',9))
plt.title('Peliculas por director')
```

```python
labels = 'H.Miyazaki', 'Takahata', 'Kondo','Morita','G.Miyazaki','Yonebayashi'
sizes = [92.7,90,91,89,62,93.5]
explode = (0.1, 0, 0, 0,0,0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
